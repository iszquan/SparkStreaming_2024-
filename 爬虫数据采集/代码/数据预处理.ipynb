```python
# 加载数据分析需要使用的库
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import datetime as dt
import warnings
```

```python
df = pd.read_csv("./boss_5.csv",sep=',',encoding='gbk')
df.head(5)
```

```python
df['city'] = df['location'].apply(lambda x: x[:2])

# Define the function to convert salary string to double
def convert_salary(row):
    salary = row['salary']
    if 'K' in salary:
        salary = salary.split('-')
        if len(salary) == 2:
            salary = (float(salary[0]) + float(salary[1].split('K')[0])) / 2 * 1000
            return salary
    return None

df['salary'] = df.apply(convert_salary, axis=1)

# 查看数据缺失情况
df.isnull().sum(axis=0)

# 统计各列空值的占比
# print(df.apply(lambda x:sum(x.isnull())/len(x),axis=0))

# 空值处理
## 删除salary为空值的行
df.dropna(subset=["salary"],inplace=True)

df.head(5)
```
